```python
#load red wine data
import pandas as pd
#red_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=';')
red_wine = pd.read_csv('/Users/macbookair/Downloads/winequality-red.csv', sep=';')
red_wine.head(5)
```

```python
#check for null values in red wine
red_wine.isnull().sum()
```

```python
red_wine.describe().T
```

```python
red_wine.shape
```

```python
# we are splitting the data in 0/1/2 
#Y = red_wine['quality'].apply(lambda value: 0 if value <= 5 else 1 if value <= 6 else 2)
# Set quality label as 0 for quality range (1-5) ; 1 for quality value 6; 2 for quality range (7-9)
red_wine['quality'] = red_wine['quality'].apply(lambda value: 0 if value <= 5 else 1 if value <= 6 else 2)
#red_wine = red_wine.drop(columns=['quality'])
#red_wine = red_wine.drop('quality', axis =1)
Y = red_wine['quality'] 
Y.head()
```

```python
X = red_wine.drop(columns=['quality'], axis=1) 
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state= 2 )
```

```python
#Standardization - 
from sklearn.preprocessing import StandardScaler
norm = StandardScaler().fit(X_train)

# transform training data
#X_train = norm.transform(X_train)
X_train_norm = norm.transform(X_train)

# transform testing data
X_test_norm = norm.transform(X_test)
#X_test = norm.transform(X_test)
```

```python
from imblearn.over_sampling import SMOTE
os = SMOTE()
X_res, Y_res = os.fit_resample(X,Y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier()
knn.fit(X_train, Y_train)
from sklearn.metrics import accuracy_score
X_test_prediction = knn.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print ('test_data_accuracy: ', test_data_accuracy)

from sklearn.metrics import classification_report
print(classification_report(Y_test, X_test_prediction, target_names=["low","medium", "high"]))

from sklearn.metrics import cohen_kappa_score
kappa = cohen_kappa_score(X_test_prediction, Y_test)
print("Cohen-Kappa score:", kappa)
```

```python
from sklearn.linear_model import LogisticRegression
# Fit the model
lr = LogisticRegression()
lr.fit(X_train, Y_train)

# Make predictions
pred_lr = lr.predict(X_test)

from sklearn.metrics import confusion_matrix
print("Confusion matrix:")
print(confusion_matrix(Y_test, pred_lr))

from sklearn.metrics import accuracy_score
print("Accuracy score:", accuracy_score(Y_test, pred_lr))

from sklearn.metrics import classification_report
print(classification_report(Y_test, pred_lr, target_names=["low","medium", "high"]))

from sklearn.metrics import cohen_kappa_score
kappa = cohen_kappa_score(pred_lr, Y_test)
print("Cohen-Kappa score:", kappa)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc= RandomForestClassifier()
rfc.fit(X_train, Y_train)
from sklearn.metrics import accuracy_score
X_test_prediction = rfc.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print ('test_data_accuracy: ', test_data_accuracy)

from sklearn.metrics import classification_report
print(classification_report(Y_test, X_test_prediction, target_names=["low","medium", "high"]))

from sklearn.metrics import cohen_kappa_score
kappa = cohen_kappa_score(X_test_prediction, Y_test)
print("Cohen-Kappa score:", kappa)
```
